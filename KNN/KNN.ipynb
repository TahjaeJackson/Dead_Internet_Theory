```python
# Library declarations
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.preprocessing import LabelEncoder
```

```python
# Load data
X_train = np.genfromtxt("./DATA/X_train_clean.csv", delimiter=',', skip_header=1, dtype=np.float64)
X_test = np.genfromtxt("./DATA/X_test_clean.csv", delimiter=',', skip_header=1, dtype=np.float64)
y_train = np.genfromtxt("./DATA/y_train.csv", delimiter=',', skip_header=1, dtype=np.float64)
y_test = np.genfromtxt("./DATA/y_test.csv", delimiter=',', skip_header=1, dtype=np.float64)
```

```python
def z_normalize(X_train, X_test):
    """
    Apply z-score normalization 
    
    Parameters:
    -----------
    X_train, X_test : numpy arrays (all data is in float format)
   
    Returns:
    --------
    X_train_normalized, X_test_normalized : numpy arrays
    """
    X_train_normalized = X_train.copy()
    X_test_normalized = X_test.copy()

    # Compute mean and std from training data ONLY
    mean = X_train.mean(axis=0)
    std = X_train.std(axis=0)

    # Avoid division by 0
    std[std == 0] = 1.0

    # Normalize both sets using TRAINING statistics
    X_train_normalized = (X_train - mean)/std
    X_test_normalized = (X_test - mean)/std

    return X_train_normalized, X_test_normalized
```

```python
X_train_normalized, X_test_normalized = z_normalize(X_train, X_test)
```

```python
from collections import Counter
```

```python
# Define functions for k-NN implementation
def l2_distance(a, b):
    """
    Compute L2 (Euclidean) distance between two vectors.
    
    Parameters:
    -----------
    a, b : numpy arrays of same shape
    
    Returns:
    --------
    distance : float
    """
    return np.sqrt(np.sum((a-b)**2))


def knn_predict(X_train, y_train, X_test, k):
    """
    Predict labels for test data using k-NN.
    
    Parameters:
    -----------
    X_train : numpy array of shape (n_train, n_features)
    y_train : numpy array of shape (n_train,)
    X_test : numpy array of shape (n_test, n_features)
    k : int, number of neighbors
    
    Returns:
    --------
    predictions : numpy array of shape (n_test,)
    """
    predictions = []

    for x_test in X_test:
        #   1. Compute distance to all training samples
        distances = np.sqrt(np.sum((X_train - x_test)**2, axis=1))
        
        #   2. Find k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices]
        
        #   3. Predict using majority voting
        vote = Counter(k_labels).most_common(1)[0][0]
        predictions.append(vote)
    
    return np.array(predictions)
    

def compute_accuracy(y_true, y_pred):
    """
    Compute classification accuracy.
    
    Returns:
    --------
    accuracy : float (between 0 and 1)
    """
    return np.mean(y_true == y_pred)
```

```python
# Try different values of k
k_values = [1, 3, 5, 7, 9]
results = []

for k in k_values:
    y_pred = knn_predict(X_train_normalized, y_train, X_test_normalized, k)
    acc = compute_accuracy(y_test, y_pred)
    results.append((k, acc))
    print(f"k={k}, Accuracy={acc:.4f}")
```
